# Principal Component Analysis

````python
pivoted.shape # (hours, days)
````

````python
pivoted.T.shape # (days, hours)
````

````python
from sklearn.decomposition import PCA
PCA(2) # two components
````

````python
pivoted.T.values
````

````python
X = pivoted.T.values
X.shape
````

````python
PCA(2).fit(X)
````

````python
X = pivoted.fillna(0).T.values
````

````python
PCA(2).fit(X)
````

````python
X2 = PCA(2).fit_transform(X)
````

````python
X2.shape
````

````python
X2
````

````python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

plt.scatter(X2[:, 0], X2[:, 1])
````

# Perform Gaussian Mixture Analysis

````python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(2) # Two clusters
gmm.fit(X)
gmm.predict(X)
````

````python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(2) # Two clusters
gmm.fit(X)
labels = gmm.predict(X)
plt.scatter(X2[:, 0], X2[:, 1], c=labels, cmap="rainbow")
plt.colorbar()
````

````python
pivoted.shape
````

````python
pivoted.T[labels == 0].T.plot(legend=False, alpha=0.1)
````

````python
pivoted.T[labels == 1].T.plot(legend=False, alpha=0.1)
````

````python
pivoted.columns
````

````python
import pandas as pd

pd.DatetimeIndex(pivoted.columns)
````

````python
pd.DatetimeIndex(pivoted.columns).dayofweek
````

````python
dayofweek = pd.DatetimeIndex(pivoted.columns).dayofweek
plt.scatter(X2[:, 0], X2[:, 1], c=dayofweek, cmap="rainbow")
plt.colorbar()
````

````python
dates = pd.DatetimeIndex(pivoted.columns)
dates[(labels == 0) & (dayofweek < 5)]
````
